```python
## Rod slicing

import numpy as np
import matplotlib.pyplot as plt


def rod(dt, T, L):
    g = 9.81
    N_t = int(round(T/dt))
    u = np.zeros(N_t+1)
    v = np.zeros(N_t+1)
    t = np.linspace(0, N_t*dt, N_t+1)
    
    # Initial condition: a 45 grados, del reposo.
    u[0] = 45*np.pi/180
    v[0] = 0
    
    # Step equations forward in time
    for n in range(N_t):
        v[n+1] = v[n] - (3/4)*(g/L)*np.cos(u[n])*dt
        u[n+1] = u[n] + v[n]*dt
    return u, v, t
    
    
def demo():
    L = 1 # Longitud del tubo
    dt = 0.01
    T = 1
    
    u, v, t = rod(dt, T, L)
    
    # Coordenada del centro de masa
    xs = L*np.cos(u)
    ys = L*np.sin(u)
    
    # Graficas de las líneas 
    X = np.linspace(0,2*L, 101)
    Y0 = -1.0*np.tan(u)[0]*X + 2*L*np.sin(u)[0] 
    Y1 = -1.0*np.tan(u)[10]*X + 2*L*np.sin(u)[10] 
    Y2 = -1.0*np.tan(u)[20]*X + 2*L*np.sin(u)[20]
    Y3 = -1.0*np.tan(u)[30]*X + 2*L*np.sin(u)[30]
    Y4 = -1.0*np.tan(u)[40]*X + 2*L*np.sin(u)[40]
    Y5 = -1.0*np.tan(u)[50]*X + 2*L*np.sin(u)[50]
    Y6 = -1.0*np.tan(u)[60]*X + 2*L*np.sin(u)[60]
    
    
    # Gráfica de la coordenada X del centro de masa
    plt.plot(t, xs, "g-")
    plt.legend(["Dependencia temporal de la coordenada x"])
    plt.show()
    
    # Graficas de la posición del CM y curvas
    for i,j in zip(xs,ys):
        plt.plot(i,j, ".", color="red")
    plt.plot(X, Y0, X, Y1, X, Y2, X, Y3, X, Y4, X, Y5, X, Y6)
    plt.legend(["Movimiento del centro de gravedad"])
    
    plt.xlim((0,2*L))
    plt.ylim((0,2*L*np.sin(np.pi/4)))
    plt.grid("on")
    plt.show()
    
demo()
```
